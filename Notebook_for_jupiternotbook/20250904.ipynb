```python
# Task 1: Create a simple dictionary that stores 2 variables, for example: first and last name.
print("Task 1:")
person = {
    'first_name': 'John',
    'last_name': 'Doe'
}
print(person)

print("-----------------------------------------------------------------------------------------------")
print("Task 2:")
# Task 2: Print out those variables stored in your previous dictionary.
print("Hello, " + person['first_name'] + " " + person['last_name'] + "!")


print("-----------------------------------------------------------------------------------------------")
print("Task 3:")
# Task3: Create a dictionary that holds 2 key: value pairs:
person['age'] = 30
person['city'] = 'New York'
print(person)


print("-----------------------------------------------------------------------------------------------")
print("Task 4:")
# Task 4: Look through your dictionary and print each pair,
for key, value in person.items():
    print(key + ": " + str(value))


print("-----------------------------------------------------------------------------------------------")
print("Task 5:")
# Task 5: Create a nested dictionary containing three dictionaries â€“ these dictionaries could be anything (favorite pets, travel locations, etc.)
favorites = {
    'pets': {
        'dog': 'Golden Retriever',
        'cat': 'Siamese',
        'bird': 'Parrot'
    },
    'travel': {
        'location1': 'Paris',
        'location2': 'Tokyo',
        'location3': 'New York'
    },
    'food': {
        'breakfast': 'Pancakes',
        'lunch': 'Sushi',
        'dinner': 'Steak'
    }
}
print(favorites)

print("-----------------------------------------------------------------------------------------------")
print("Task 6:")
# Task 6: Loop through the dictionaries and print a message for each.
for category, items in favorites.items():
    print("My favorite " + category + " are:")
    for key, value in items.items():
        print(" - " + key + ": " + value)
    print()
```
